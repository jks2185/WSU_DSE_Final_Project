```python
import pandas as pd

# Load the CSV file into a Pandas dataframe
df = pd.read_csv('Data.csv')

# Print the first 5 rows of the dataframe
print(df.head())
```

```python
#Graph 1
#Year on Year Average Income Comparison between Male and Female Population
import pandas as pd 
import plotly.express as px

#convert SEX column to str type
df['SEX'] = df['SEX'].astype(str)

#replace 1 with M and 2 with F
df['SEX'] = df['SEX'].replace({'1': 'M', '2': 'F'})
df.dropna(subset=['SEX'], inplace=True)

#drop rows where SEX is blank or not M or F
df = df[df['SEX'].isin(['M', 'F'])]

#convert INCOME column to numeric type
df['INCOME'] = pd.to_numeric(df['INCOME'], errors='coerce')

#calculate year on year average income by sex category
pivot = df.pivot_table(index=['YYYY'], columns='SEX', values='INCOME', aggfunc='mean')

#reset the index
pivot = pivot.reset_index()

#create double bar graph using plotly express
fig = px.bar(pivot, x='YYYY', y=['M', 'F'], barmode='group')

#set plot title and labels
fig.update_layout(title='Year on Year Average Income by Sex and Category', xaxis_title='Year, Sex', yaxis_title='Average Income')

#display the plot
fig.show()

#Print conclusion
print("The above graph displays Year on year avergae income for Male and Female. The difference between the average income between both the sex varied over the years." "We can conclude that in 1979 has the least difference where as 2010 has the highest variation.")
```

```python
#Graph 2
# Correlation between Inflation and buying attitude for Household Items, House and Vehicle

import pandas as pd
import plotly.express as px
import numpy as np

# select columns for correlation plot
corr_columns = ['PX1Q1', 'DUR', 'HOM', 'CAR']
corr_df = df[corr_columns]

# calculate correlation matrix
corr_matrix = corr_df.corr()

# create annotation list
annotations = np.round(corr_matrix.values, 2)

# plot heatmap
fig = px.imshow(corr_matrix.values,
                x=corr_columns, y=corr_columns,
                color_continuous_scale='blues',
                zmin=-1, zmax=1,
                labels=dict(x='', y=''))

# add annotations to the heatmap
for i in range(len(corr_columns)):
    for j in range(len(corr_columns)):
        fig.add_annotation(x=corr_columns[i], y=corr_columns[j],
                           text=str(annotations[i][j]),
                           showarrow=False, font=dict(color='white'))

# update layout to show the color scale and adjust the margins
fig.update_layout(title="Correlation Matrix",
                  coloraxis_colorbar=dict(title='Correlation'),
                  margin=dict(l=100, r=100, t=50, b=100))

# show the figure
fig.show()

#Conclusion
print("There is positive correlation between Inflation and buying attitude of people. People are still willing to buy household items, house and vehicles even when inflation is increasing.")
```

```python
#Graph 3
# Average Income of people who had college degree

import pandas as pd
import plotly.express as px

# Count the number of respondents who are college graduates and those who are not
counts = df['ECLGRD'].value_counts()

# Get the sizes and labels for the pie chart
sizes = [counts[0], counts[1]]
labels = ['Non-College Graduates', 'College Graduates']

# Create the pie chart
fig = px.pie(values=sizes, names=labels, title='Proportion of College Graduates in Respondents')
fig.update_traces(textposition='inside', textinfo='percent+label')

# Display the chart
fig.show()

# Convert income column to numeric
df['INCOME'] = pd.to_numeric(df['INCOME'], errors='coerce')

# Replace blanks with 0
df['INCOME'] = df['INCOME'].fillna(0)

# calculate average income for values 1 and 5 in ECLGRD column
avg_income_1 = df.loc[df['ECLGRD'] == '1', 'INCOME'].mean()
avg_income_5 = df.loc[df['ECLGRD'] == '5', 'INCOME'].mean()

print("\033[1mAverage income for College graduates:\033[0m $", round(avg_income_1, 2))
print("\033[1mAverage income for Non-college graduates:\033[0m $", round(avg_income_5, 2))
print("")
print("The average income of people with college degree is much higher when compared to the ones with no college degree.")
```

```python
#Graph 4
#Comaprision of how people think they are financially currently than a year ago and how it will be a year from now

import pandas as pd
import plotly.graph_objects as go

df = df[(df["PAGO"] != 9) & (df["PEXP"] != 9)]
df = df[df["PAGO"].isin([1, 3, 5, 8])]
df = df[df["PEXP"].isin([1, 3, 5, 8])]

df_count_pexp = df["PEXP"].value_counts().sort_index()
df_count_pago = df["PAGO"].value_counts().sort_index()

fig = go.Figure()

fig.add_trace(go.Bar(x=[1, 3, 5, 8], y=df_count_pago, name="Current from a Year Ago", marker_color='#1f77b4'))
fig.add_trace(go.Bar(x=[1, 3, 5, 8], y=df_count_pexp, name="Year Ahead", marker_color='#ff7f0e'))

fig.update_layout(title="Financial position people think currently from a year ago and a year ahead", xaxis_title="Category", yaxis_title="Count",
                  xaxis=dict(
                      tickmode='array',
                      tickvals=[1, 3, 5, 8],
                      ticktext=['Better', 'Same', 'Worst', 'Not Sure']
                  ))

fig.show()

print("We can conclude from above graph that majority of the people over the years think they are currently financially better than they were a year ago and would be same in next one year.")
```

```python
#Graph 5
#Showing the ratio of people who think its good time to buy household item, house, car and sellign house each year

import pandas as pd
import plotly.express as px

# Convert column values to numeric
df[['HOM', 'DUR', 'SHOM', 'CAR']] = df[['HOM', 'DUR', 'SHOM', 'CAR']].apply(pd.to_numeric)

# Create a pivot table
pivot_table = pd.pivot_table(df, index=['YYYY'], values=['HOM', 'DUR', 'SHOM', 'CAR'], aggfunc=lambda x: sum(x==1))

# Create the stacked bar graph
fig = px.bar(pivot_table, barmode='stack')

#Set the legend titles directly using the 'name' property of each trace
fig.for_each_trace(lambda t: t.update(name=t.name.replace('DUR', 'Household Item').replace('HOM', 'Home').replace('CAR', 'Vehicle').replace('SHOM', 'Selling the house')))

# Add axis titles and legend
fig.update_layout(title='Ratio of people who think its good to buy household item, house, car and sellign house each year', xaxis_title='Year', yaxis_title='Count', legend_title='Key')
                

# Show the plot
fig.show()

#Conclusion
print("Based on the data provided, we can conclude that buying attitude towards household items has generally increased over the years. However, there has been a decline on buying and selling homes since the early 1990s. Buying cars has been relatively stable over the years, with some fluctuations. The stack bar graph shows that the majority is on household items, followed by cars, and then buying and selling homes.It is also worth noting that there is a sharp decrease in the year 2022, followed by 2023.")
```
